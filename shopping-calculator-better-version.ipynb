```python
user_input= input("Enter the prices (eg:10,20,30): ")
discount_type= input("Enter the discount type (normal(n)/percent(%)/none): ")
discount= input("Enter the discount (if none type 0): ")
total= 0
text_prices= user_input.split(",")
for items in text_prices:
    total+= int(items)
if discount_type== "n":
    total-= int(discount)
elif discount_type== "%":
    total= total* (1-(int(discount)/100))
else:
    pass
if int(discount)== "0":
    pass
print(f"Your total is {total}")
```
